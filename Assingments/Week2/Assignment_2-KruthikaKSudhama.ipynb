```python
import pandas as pd
```

```python
data = pd.read_csv('adult_data.csv')
data.head(20)
```

### 1. How many men and women are represented in this dataset?

```python
female_count =int(data[['sex']][data['sex'] == 'Female'].count())
male_count =int(data[['sex']][data['sex'] == 'Male'].count())
print(f"Female count = {female_count}\n Male count = {male_count}")

```

### 2. What is the average age of women?

```python
data['age'][data.sex == 'Female'].mean()
```

### 3. What is the percentage of german citizens?

```python
nt = float(data[['native-country']][data['native-country'] == 'Germany'].count())
rows = data.shape[0]
print(nt*100/rows)
```

### 4. What is the mean of age for those who earn more than 50K per year and those who earn less than 50K per year? 

```python
less_than_50K = int(data[['age']][data.salary == '<=50K'].count())
more_than_50K = data.shape[0] - less_than_50K
print(f"<=50K = {less_than_50K}, >=50K = {more_than_50K}")
```

### 5. What is the standard deviation of age for those who earn more than 50K per year and those who earn less than 50K per year? 

```python
data['age'].fillna(0, inplace=True)
std_more = float(data[['age']][data.salary != '<=50K'].std(axis = 0))
std_less = float(data[['age']][data.salary == '<=50K'].std(axis = 0))
print(f"standard deviations of ages of those with more than 50K salary and less than 50K salary are {std_more}, {std_less} respectively")
```

### 6. Is it true that people who earn more than 50K have at least high school education?

```python

data.rename(columns = {'education-num' : 'education_num'}, inplace = True)
salary_ = data.groupby('salary')
more_salary = salary_.get_group('>50K')

ans = int(more_salary[['education_num']][more_salary.education_num < 9].count())
if ans>0 :
    print("False")
```

### 7. Display age statistics for each race and each gender. Use groupby() and describe(). Find the maximum age of men of Amer-Indian-Eskimo race.

```python
race_ = data.groupby('race')
gender = data.groupby('sex')
print(f"Race age statistics \n {race_.age.describe()} \n \n Gender age statistics \n {gender.age.describe()}")
print(f"\n\nMaximum age of Amer-Indian-Eskimo race is {race_.get_group('Amer-Indian-Eskimo').age.max()}")


```

### 8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men? Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.

```python




data.rename(columns = {'marital-status' : 'marital_status'}, inplace = True)
salary_ = data.groupby('salary')
more_salary = salary_.get_group('>50K')
sex_ = more_salary.groupby('sex')
male = sex_.get_group('Male')

mar_status = int(male[['marital_status']][male.marital_status.str.slice(start = 0, stop = 7) == 'Married'].count())
unmarried = male.shape[0] - mar_status
ans = "Unmarried men"
if mar_status >unmarried :
    ans = "Married men"
print(f"Married men: {mar_status}\n Single men: {unmarried}\n\n Therefore the proportion of how earn alot is more among - {ans}")
```

### 9. What is the maximum number of hours a person works per week? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?

```python

hours_per_week = data.groupby('hours-per-week')
max_hours = hours_per_week.get_group(data['hours-per-week'].max())
num_max_hours = max_hours.shape[0]
print(f"Maximum num of working hours per week : {data['hours-per-week'].max()} \nNumber of people working for maximum hours is {num_max_hours}")
earn_lot = int(max_hours[['salary']][max_hours.salary == '>50K'].count())
print(f"\nPercentage of those who earn a lot - {earn_lot*100/ num_max_hours}")


```

### 10. Count the average time of work for those who earn a little and a lot for each country. What will these be for Japan?

```python
data.rename(columns = {'hours-per-week' : 'hours_per_week', 'native-country' : 'native_country'}, inplace = True)

salary_ = data.groupby('salary')
more_salary = salary_.get_group('>50K')
less_salary = salary_.get_group('<=50K')
less_countries = less_salary.groupby('native_country')
more_countries = more_salary.groupby('native_country')
less_japan = more_countries.get_group('Japan')
more_japan = less_countries.get_group('Japan')
print(f"Average time of work of those who earn a lot {more_countries.hours_per_week.mean()}\nAverage time of work of those who earn less {less_countries.hours_per_week.mean()}\n\nAverage time of work of people of Japan \nWho earn alot is {more_japan.hours_per_week.mean()}\nWho earn less is {less_japan.hours_per_week.mean()}")


```
